```python
# Importing the required library 'pandas'.
# pandas is a library which is helpful for handling  tables and reading files
# this library contains number of functions that can be useful to read data from tables or files
# we set an alias name 'pd' for pandas

import pandas as pd
```

```python
# Using a function "read_csv" from pandas library inorder to fetch the data from csv file (dataset)
# Fetched datas are storing in a variable "dataset"

dataset = pd.read_csv("Salary_Data.csv")
```

```python
# YearsExperience - independent variable
# Salary - dependent variable
# Below data seems preprocessed dataset

dataset
```

```python
# Fetch 'YearsExperience' column values from csv file and assign it in 'independent' variable

independent= dataset[["YearsExperience"]]
```

```python
# Fetch 'Salary' column values from csv file  and assign it in 'dependent' variable

dependent= dataset[["Salary"]]
```

```python
# To save memory, here we are calling only a particular function 'train_test_split'  from sk learn library instead of importing the whole library
# It is used to split train and test dataset
# This function 'train_test_split' returns four output values and storing those values in four variables
# Splitting the dataset in the ratio of 70:30,  70 for train dataset and 30 for test dataset
# x means input ; y means output

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# sk learn are providing all the machine learning algorithms 
# linear model is the function from where we are importing the LinearRegression
# Using this function, we are going to create MODEL.
# 'Learning Phase' is getting happend here in this block of codes
# While using any algorthim for model creation, Only the below block will be changed 

from sklearn.linear_model import LinearRegression

# Creating a variable 'regressor'
# LinearRegression is a class where its all functions are assigned to the variable 'regressor'
regressor = LinearRegression() 

# passing the two parameters 'X_train' and 'Y_Train' as input data to  function 'fit'.
# After the execution of below line, 'weight' and 'bias' are calculated which means that the required MODEL 'y=mx+c' is ready
regressor.fit(X_train,Y_train)

```

```python
# coef_ is a method that returns the calculated value of weight and it is assigned to the variable 'weight'
weight=regressor.coef_
weight
```

```python
# intercept_ is a method that returns the calculated value of bais and it is assigned to the variable 'bais'
bais=regressor.intercept_
bais
```

```python
# After Model creation, Evaluation metrics needs to be takesplace ( Evaluation metric means cross checking the created MODEL)
# Evaluation metrics has to be taken for test data
Y_predict = regressor.predict(X_test)
```

```python
# In evaluation metrixs, we are going to create 'R square' value
from sklearn.metrics import r2_score
r_score = r2_score(Y_test,Y_predict)
```

```python
r_score
```

```python
# We are importing a function called 'pickle' which is used here to save the created MODEL
# sav is the file extension for pickle
#  Only the filename is created  and the MODEL is not saved yet

import pickle
filename="finalized_model_linear.sav"
```

```python
#  Below line is for - saving the evaluated MODEL 
# 'wb' means write binary data  

pickle.dump(regressor,open(filename,'wb'))
```

```python
# After saved the MODEL, we need to load it.
# Below line shows how to load the MODEL
# 'rb' means read binary data

loaded_model = pickle.load(open("finalized_model_linear.sav",'rb'))

# After loaded the MODEL, we are going to test it for real time check
# Applying the input value or Number of experience as '13' that needs to be passed to 'predict' method and obtain the output value or Salary 
result = loaded_model.predict([[13]])

# Finally, we are able to predict the output value based on the input data applied. 
result
```
